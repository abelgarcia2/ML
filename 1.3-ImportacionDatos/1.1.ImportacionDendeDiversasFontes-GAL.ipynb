## 1. Consulta de datasets precargados
Existen conxuntos de datos dentro da librería scikit-learn que podemos cargar de forma simple para facer probas.

````python
from sklearn import datasets

iris_data = datasets.load_iris()
iris_data
````

````python
iris = pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
iris['species'] = pd.Series(iris_data.target_names[iris_data.target])
iris.head()
````

````python
# Gráfico de puntos
iris.plot.scatter(x='petal length (cm)',
                  y='petal width (cm)')
````
